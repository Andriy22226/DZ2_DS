```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
date = pd.date_range(start='2021-09-01', freq='D', periods=8)
plt.figure(figsize=(15,6))
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20])
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
plt.xlabel('', fontsize='small', color='midnightblue')
plt.xticks(fontsize=5)
plt.yticks(fontsize=10)
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  . ', fontsize=15, fontweight=100)
plt.text(date[0], 20, '  ', color="blue")
plt.legend()
plt.show()
```

```python
date = pd.date_range(start='2021-09-01', freq='D', periods=8)
fig, axs = plt.subplots()
axs.plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs.plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')
plt.xlabel('', fontsize='small', color='midnightblue')
plt.ylabel('', fontsize='small', color='midnightblue')
plt.title('  . ', fontsize=15)
plt.text(date[0], 15, '  ', color="blue")
plt.legend()
plt.show()
```

```python
#    

date = pd.date_range(start='2021-09-01', freq='D', periods=8)
#plt.figure(figsize=(30,50))
fig, axs = plt.subplots(2, 1)
fig.tight_layout(pad=5.0)

axs[0].plot(date, [23, 17, 17, 16, 15, 14, 17, 20], label='day temperature')
axs[1].plot(date, [19, 11, 16, 11, 10, 10, 11, 16], label='night temperature')

axs[0].set_title('', fontsize=10)
axs[1].set_title('', fontsize=10)

fig.suptitle('  . ', fontsize=15)

plt.show()
```

```python
# :   ,    linestyle

date = pd.date_range(start="2021-09-01", freq="D", periods=8)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle=":",
    color="#061358",
)
plt.xlabel("", fontsize="small", color="midnightblue")
plt.ylabel("", fontsize="small", color="midnightblue")
plt.title("  . ", fontsize=15)
plt.legend()
plt.show()
```

```python
#  ,    
date = pd.date_range(start="2021-09-01", freq="D", periods=8)
plt.plot(
    date,
    [23, 17, 17, 16, 15, 14, 17, 20],
    label="day temperature",
    linestyle="--",
    color="#FF5733",
    linewidth=2,
    marker="D",
)
plt.plot(
    date,
    [19, 11, 16, 11, 10, 10, 11, 16],
    label="night temperature",
    linestyle=":",
    color="#061358",
    linewidth=2,
    marker="^",
)
plt.ylim(0, 25)
plt.xlabel("", fontsize="small", color="midnightblue")
plt.ylabel("", fontsize="small", color="midnightblue")
plt.title("  . ", fontsize=15)
plt.legend()
plt.grid(color='r', linestyle='-', linewidth=2)
plt.show()
```

```python
#  

labels = [
    "Junior Software Engineer",
    "Senior Software Engineer",
    "Software Engineer",
    "System Architect",
    "Technical Lead",
]

data = [63, 31, 100, 2, 11]
explode = [0.15, 0.2, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.3f%%",
    pctdistance=1.15,
    labeldistance=1.35,startangle=-50
)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

theta_max = 8 * np.pi
n = 1000
theta = np.linspace(0, theta_max, n)
x = theta
z = np.sin(theta)
y = np.cos(theta)
ax.plot(x, y, z, "g")

plt.show()
```

```python
#  

fig = plt.figure()
ax = fig.add_subplot(projection="3d")

x = [5, 10, 15, 20]
z = [10, 0, 5, 15]
y = [0, 10, 5, 25]
s = [150, 130, 30, 160]
ax.scatter(x, y, z, s=s)

plt.show()
```

```python
#  
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_wireframe(x, y, z)

plt.show()
```

```python
# 
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_surface(x, y, z)

plt.show()
```
